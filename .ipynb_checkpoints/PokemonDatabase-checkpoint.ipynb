```python
#Creating connection to database, storing metadata and the connection into variables
engine = create_engine(f'postgresql+psycopg2://{username}:{password}@localhost:{port}/Pokemon')

meta = MetaData()

connection = engine.connect()
```

```python
df.dtypes
```

```python
#Creating the SQL Table
pokemon = Table(
    'Pokemon', meta,
    Column('id', Integer, primary_key=True),
    Column('Name', String),
    Column('Type', String),
    Column('Type1', String),
    Column('Type2', String),
    Column('HP', Integer),
    Column('Atk', Integer),
    Column('Def', Integer),
    Column('SpAk', Integer),
    Column('SpDf', Integer),
    Column('Spd', Integer),
    Column('Total', Integer),
    Column('Generation', Integer)
)
meta.create_all(engine)
```

```python
print(engine.table_names())
```

```python
sql_insert_query = """INSERT INTO Pokemon (id, Name, Type, Type1, Type2, HP, 
Atk, Def, SpAk, SpDf, Spd, Total, Generation)
VALUES ({df})"""
cusor = connection.cursor()
cusor.execute(sql_insert_query)
connection.commit()
```
